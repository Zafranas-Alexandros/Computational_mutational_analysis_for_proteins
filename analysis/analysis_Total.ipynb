This is a notebook to be used to create plots based on cumulative information from the six(6) plain MD simulations performed.

```python
#load necessary modules and libs
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pathlib import Path
import os
import pandas as pd
import plotly.express as px 
from matplotlib.colors import ListedColormap
import math
```

```python
#Use numpy to load the data from all files  -- RGYR
t, rgyr_wt = np.loadtxt("../TP53_wt_Zn2/data_wt_Zn2/Analysis_zaf/rgyr.dat", unpack=True)
t, rgyr_y220d = np.loadtxt("../TP53_y220d_Zn2/data_y220d_Zn2/Analysis_zaf/rgyr.dat", unpack=True)
t, rgyr_y220c = np.loadtxt("../TP53_y220c_Zn2/data_y220c_Zn2/Analysis_zaf/rgyr.dat", unpack=True)
t, rgyr_p1 = np.loadtxt("../TP53_y220p1_Zn2/data_y220p1_Zn2/Analysis_zaf/rgyr.dat", unpack=True)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

# Plot lines with transparency and labels
ax.plot(t, rgyr_wt, linewidth=0.3, marker='o', markersize=2, alpha=0.7, color='blue', label="WT")
ax.plot(t, rgyr_y220d, linewidth=0.3, marker='s', markersize=2, alpha=0.7, color='orange', label="Y220d")
ax.plot(t, rgyr_y220c, linewidth=0.3, marker='^', markersize=2, alpha=0.7, color='green', label="Y220c")
ax.plot(t, rgyr_p1, linewidth=0.3, marker='*', markersize=2, alpha=0.7, color='purple', label="Y220p1")

# Axis labels
ax.set_xlabel('Frame Number')
ax.set_ylabel(r'radius of gyration')

# Add a legend
ax.legend()

# Grid for better readability
plt.grid(linewidth=0.3, linestyle='-')
#plt.show()


#save the figure
plt.savefig('RMSG_all2all_Zn2')
```

```python
#Use numpy to load the data from all files  -- RMSD
t, rmsd_wt = np.loadtxt("../TP53_wt_Zn2/data_wt_Zn2/Analysis_zaf/rmsd.dat", unpack=True)
t, rmsd_y220d = np.loadtxt("../TP53_y220d_Zn2/data_y220d_Zn2/Analysis_zaf/rmsd.dat", unpack=True)
t, rmsd_y220c = np.loadtxt("../TP53_y220c_Zn2/data_y220c_Zn2/Analysis_zaf/rmsd.dat", unpack=True)
t, rmsd_p1 = np.loadtxt("../TP53_y220p1_Zn2/data_y220p1_Zn2/Analysis_zaf/rmsd.dat", unpack=True)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

# Plot lines with transparency and labels
ax.plot(t, rmsd_wt, linewidth=0.3, marker='o', markersize=2, alpha=0.7, color='blue', label="WT")
ax.plot(t, rmsd_y220d, linewidth=0.3, marker='s', markersize=2, alpha=0.7, color='orange', label="Y220d")
ax.plot(t, rmsd_y220c, linewidth=0.3, marker='^', markersize=2, alpha=0.7, color='green', label="Y220c")
ax.plot(t, rmsd_p1, linewidth=0.3, marker='*', markersize=2, alpha=0.7, color='purple', label="Y220p1")

# Axis labels
ax.set_xlabel('Frame Number')
ax.set_ylabel(r'RMSD [$\AA$]')

# Add a legend
ax.legend()

# Grid for better readability
plt.grid(linewidth=0.3, linestyle='-')
#plt.show()

#save the figure
plt.savefig('RMSD_all2all_Zn2')
```

```python
#Use numpy to load the data from all files  -- RMSD
rmsf_wt = np.loadtxt("../TP53_wt_Zn2/data_wt_Zn2/Analysis_zaf/rmsf.dat", unpack=True)
rmsf_y220d = np.loadtxt("../TP53_y220d_Zn2/data_y220d_Zn2/Analysis_zaf/rmsf.dat", unpack=True)
rmsf_y220c = np.loadtxt("../TP53_y220c_Zn2/data_y220c_Zn2/Analysis_zaf/rmsf.dat", unpack=True)
rmsf_p1 = np.loadtxt("../TP53_y220p1_Zn2/data_y220p1_Zn2/Analysis_zaf/rmsf.dat", unpack=True)

t = list(range(94,len(rmsf_wt)+94))
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

# Plot lines with transparency and labels
ax.plot(t, rmsf_wt, linewidth=0.3, marker='o', markersize=2, alpha=0.8, color='blue', label="WT")
ax.plot(t, rmsf_y220d, linewidth=0.3, marker='s', markersize=2, alpha=0.8, color='orange', label="Y220d")
ax.plot(t, rmsf_y220c, linewidth=0.3, marker='^', markersize=2, alpha=0.8, color='green', label="Y220c")
ax.plot(t, rmsf_p1, linewidth=0.3, marker='*', markersize=2, alpha=0.8, color='purple', label="Y220p1")

# Axis labels
ax.set_xlabel('Residue Index')
ax.set_ylabel(r"C$_\alpha$ RMSF [$\AA$]")

# Add a legend
ax.legend()

# Grid for better readability
plt.grid(linewidth=0.3, linestyle='-')
#plt.show()

#save the figure
plt.savefig('RMSF_all2all_Zn2')
```
